A substring of s

The location of a substring s[j:k]

```python
def SUBS_motif (string_s, string_t):

  pattern = '(?=ATAT)'
  positions = [match.start() + 1 for match in re.finditer(pattern, string_s)]
  #Find string_t in string_s

  #In re.finditer, first arg = pattern to search for and second arg = string to search through
  return positions
```
