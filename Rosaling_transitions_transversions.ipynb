Problem
For DNA strings s1
 and s2
 having the same length, their transition/transversion ratio R(s1,s2)
 is the ratio of the total number of transitions to the total number of transversions, where symbol substitutions are inferred from mismatched corresponding symbols as when calculating Hamming distance (see “Counting Point Mutations”).

Given: Two DNA strings s1
 and s2
 of equal length (at most 1 kbp).

Return: The transition/transversion ratio R(s1,s2)
.

```python
#String 1: ATGCTGACTGACCCATGGA
#String 2: TGATGGTACCGTAGTCAGG

#Return: # Transitions/# Transversions

#Purines: A,G
#Pyrimidines: T,C

#Transitions DICTIONARY: T -> C, C ->T, A -> G, G -> A
#Transversions DICTIONARY: T -> A, T -> G, C -> A, C -> G, A -> T, A -> C, G -> T, G -> C

#Idea:
#Start with transition = 0, and transversion = 0
#For nucleotide in string 1, if transition[nucleotide] exists, then transitions + 1
#else: if transversions[nucleotides] exists, then transversion +=1,
#else next
```

```python
import numpy as np
import pandas as pd
```

```python
transition = {'T':'C', 'C':'T', 'A':'G', 'G':'A'}
transversion = {'T':'A', 'T':'G', 'C':'A', 'C':'G', 'A':'T', 'A':'C', 'G':'T', 'G':'C'}

#We could use a dictionary method if we want
```

```python
def transisitions_transversion_ratio(sequence1, sequence2):

  transitions_count = 0
  transversions_count = 0

  purines = {'A', 'G'}
  pyrimidines = {'C', 'T'}

  for i in range(len(sequence1)):
    n1 = sequence1[i]
    n2 = sequence2[i]

    if n1 != n2: # Only count if there's a mismatch
      if (n1 in purines and n2 in purines) or \
         (n1 in pyrimidines and n2 in pyrimidines):
        transitions_count += 1
      elif (n1 in purines and n2 in pyrimidines) or \
           (n1 in pyrimidines and n2 in purines):
        transversions_count += 1

  if transversions_count == 0:
    # If there are transitions but no transversions, return the specific string as per original logic.
    # If there are no mismatches at all (both counts are 0), this will also return "No transversions".
    return "No transversions"
  else:
    ratio = transitions_count / transversions_count
  return ratio
```
